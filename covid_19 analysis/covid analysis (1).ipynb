# Model Building - SARIMAX

**Model Building**

***ARIMA***
ARIMA (AutoRegressive Integrated Moving Average) and SARIMA (Seasonal AutoRegressive Integrated Moving Average) are two popular models for analyzing and forecasting time series data.

ARIMA is a statistical model that uses the past values of a time series to predict its future values. It has three components: autoregression (AR), differencing (I), and moving average (MA).

AR component: The autoregressive component captures the linear relationship between the current observation and the past observations. (q)

I: The differencing component removes the stationarity from the time series. (d)

MA: The moving average component models the error term as a linear combination of past error terms.(p)
better version of arima is SARIMAX

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model = SARIMAX(train,order = (1,1,2), seasonal_order = (1,1,2,7))
model = model.fit()
```

```python
#predictions
confirmed['confirmed_prediction']=model.predict(start = len(train), end = len(train)+len(test)-1 )
confirmed[['y','confirmed_prediction']].plot()
```

## forcasting on confirmed case

```python
from prophet import Prophet
```

```python
m = Prophet(interval_width = 0.95)
m.fit(confirmed)
```

```python
future = m.make_future_dataframe(periods = 7,freq = 'D')
future.tail(7)
```

```python
forecast = m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail(7)
```

```python
confirmed_plot = m.plot(forecast)
```
