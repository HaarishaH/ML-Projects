# Data Visualisation

```python
import plotly.graph_objects as go
data = {'Class': ['M', 'R'], 'Count': [53, 46]}
df = pd.DataFrame(data)
df['Proportion'] = (df['Count'] / df['Count'].sum()) * 100 
fig = go.Figure()

fig.add_trace(go.Bar(
    x=df['Class'],             
    y=df['Count'],            
    marker=dict(color=['pink', 'skyblue']), 
    text=df['Proportion'].apply(lambda x: f'{x:.2f}%'),  
    textposition='outside'   
))
fig.update_layout(
    title="Class Counts with Proportions",
    xaxis=dict(title="Class", tickmode='linear'),
    yaxis=dict(title="Count"),
    font=dict(size=13),
    height=600,
    width=800
)
fig.show()
```

# Separating the features and labels

```python
x = d.drop(columns= 60, axis = 1)
y = d[60]
```

```python
x
```

```python
y
```

# Train test split

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size = 0.2, stratify = y, random_state = 31)
```

```python
print(x.shape, x_train.shape,x_test.shape)
```

# Model training

## Logistic Regression

```python
log = LogisticRegression()
log.fit(x_train,y_train)
```

# Model evaluation

```python
x_train_pred = log.predict(x_train)
train_accuracy = accuracy_score(x_train_pred,y_train)
train_accuracy
```

```python
x_test_pred = log.predict(x_test)
test_accuracy = accuracy_score(x_test_pred,y_test)
test_accuracy
```

```python
def con(x_train_pred,y_train,dom):
    print(f'{dom} confusion matrix')
    cf = confusion_matrix(x_train_pred,y_train)
    sns.heatmap(cf,annot = True, cmap = 'pink')
    plt.tight_layout()
    plt.show()
```

```python
con(x_train_pred,y_train,dom = 'Train Data')
```

```python
con(x_test_pred,y_test,dom = 'Test Data')
```
